```python
%pylab inline
from commpy.filters import rrcosfilter

#sample_rate = 48000
sample_rate = 240000 
#times 5 since we are 5x the original codec bitrate and this gets us into the ballpark
samples_per_symbol = 10
symbol_rate = sample_rate/samples_per_symbol


def response_plot(taps, sample_rate, legend):

    duration = len(taps) / sample_rate
    t = np.linspace(0, duration, len(taps), endpoint=True)
    plt.figure()
    plt.rcParams['figure.figsize'] = [12, 4]
    ax = plt.subplot(1, 1, 1)
    plt.xlabel('Time (msec)')
    plt.locator_params(axis='y', nbins=8)
    plt.grid()
    #plt.xticks(np.arange(0, duration, 1.0/4800.0), rotation=45) #original, symbol rate?
    plt.xticks(np.arange(0, duration, 1.0/(symbol_rate)), rotation=90)

    plt.plot(t, taps, '-o')
    plt.legend(legend)

#taps = rrcosfilter(int(40 * 8), 0.5, 1.0, 40.0)[1][1:] #original

#Parameters:
#N (int) – Length of the filter in samples. Original was 320, expressed as 40 times 8. 
# I think that the samples per symbol is 10. We can go with a span of 8 symbols. 
# 6 is commonly used, but let's go with 8 for now. 
# The length of the filter would then be 80. 
#alpha (float) – Roll off factor (Valid values are [0, 1]). We have 0.5. That's fine. Let's keep it.
#Ts (float) – Symbol period in seconds. People pretend it's 1 to make it easy. We use 1/symbol_rate
#Fs (float) – Sampling Rate in Hz. We can use sample_rate directly. 

taps = rrcosfilter(int(10 * 8), 0.5, 1/symbol_rate, sample_rate)[1][1:]

response_plot(taps, sample_rate, ["Erroneous Filter Resoponse"])
```

```python
gain = sum(taps)
print(gain)
print(max(taps))
print(32767 // max(taps))

print(",\n".join([str(int(t * 32767 // max(taps))) for t in taps]))
```
